```python
# ========== Задание 1: Классификация 5 классов (Reuters + LSTM) ==========
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'   # скрыть предупреждения
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'  # отключить oneDNN (устраняет сообщение AVX-512)

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers, models
from tensorflow.keras.datasets import reuters
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical


# Параметры
NUM_WORDS = 10000   # словарь
MAX_CLASSES = 5     # оставим 5 классов: метки 0..4
EMBED_DIM = 64
LSTM_UNITS = 64
BATCH_SIZE = 128
EPOCHS_BASELINE = 10
EPOCHS_LSTM = 8
VALIDATION_SPLIT = 0.1
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# 1) Загрузка и фильтрация по классам (оставляем 5 первых классов)
(train_data, train_labels), (test_data, test_labels) = reuters.load_data(num_words=NUM_WORDS)
mask_train = train_labels < MAX_CLASSES
mask_test = test_labels < MAX_CLASSES
train_data = np.array(train_data)[mask_train]
train_labels = np.array(train_labels)[mask_train]
test_data = np.array(test_data)[mask_test]
test_labels = np.array(test_labels)[mask_test]

print("Samples (train/test):", len(train_data), len(test_data))
print("Example doc (token ids):", train_data[0][:20])
print("Label example:", train_labels[0])

# 2) Векторизация для базовой dense модели (one-hot bag-of-words)
def vectorize_sequences(sequences, dimension=NUM_WORDS):
    results = np.zeros((len(sequences), dimension), dtype='float32')
    for i, seq in enumerate(sequences):
        results[i, np.array(seq)] = 1.0
    return results

x_train_dense = vectorize_sequences(train_data)
x_test_dense = vectorize_sequences(test_data)
y_train = to_categorical(train_labels, NUM_WORDS)  # we'll reassign correct dim below
y_train = to_categorical(train_labels, MAX_CLASSES)
y_test = to_categorical(test_labels, MAX_CLASSES)

# Отделим валидацию
val_size = int(0.1 * x_train_dense.shape[0])
x_val_dense = x_train_dense[:val_size]
partial_x_train_dense = x_train_dense[val_size:]
y_val = y_train[:val_size]
partial_y_train = y_train[val_size:]

# ========== Baseline: полносвязанная модель ==========
baseline_model = keras.Sequential([
    layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l1(1e-3), input_shape=(NUM_WORDS,)),
    layers.Dropout(0.5),
    layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l1(1e-3)),
    layers.Dropout(0.5),
    layers.Dense(MAX_CLASSES, activation='softmax')
])
baseline_model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
baseline_model.summary()

t0 = time.time()
history_base = baseline_model.fit(partial_x_train_dense, partial_y_train,
                                  epochs=EPOCHS_BASELINE, batch_size=BATCH_SIZE,
                                  validation_data=(x_val_dense, y_val), verbose=2)
t_base = time.time() - t0
print("Baseline training time (s):", t_base)
res_base = baseline_model.evaluate(x_test_dense, y_test, verbose=0)
print("Baseline test loss/acc:", res_base)

# ========== Prepare data for LSTM model ==========
# Для LSTM используем Embedding + pad_sequences + masking
# Сначала оставим последовательности в "индексном" виде.
# Разделяем валидацию заново по индексным последовательностям:
x_val_seq = train_data[:val_size]
partial_x_train_seq = train_data[val_size:]
y_val = to_categorical(train_labels[:val_size], MAX_CLASSES)
partial_y_train = to_categorical(train_labels[val_size:], MAX_CLASSES)

# Паддим по длине (вариант A: pad до максимум в валидации/трейне отдельно)
padded_x_val = pad_sequences(x_val_seq, padding='post')   # pad до длины наибольшей в валидации
padded_partial_x_train = pad_sequences(partial_x_train_seq, padding='post',
                                       maxlen=padded_x_val.shape[1])  # pad to same length

maxlen_global = padded_x_val.shape[1]
print("maxlen used for initial LSTM pad:", maxlen_global)

# ========== LSTM модель (без упорядочивания) ==========
lstm_model = keras.Sequential([
    layers.Embedding(NUM_WORDS, EMBED_DIM, input_length=maxlen_global, mask_zero=True),
    layers.Bidirectional(layers.LSTM(LSTM_UNITS, dropout=0.1, recurrent_dropout=0.1)),
    layers.Dense(MAX_CLASSES, activation='softmax')
])
lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
lstm_model.summary()

t0 = time.time()
history_lstm = lstm_model.fit(padded_partial_x_train, partial_y_train,
                              epochs=EPOCHS_LSTM, batch_size=BATCH_SIZE,
                              validation_data=(padded_x_val, y_val), verbose=2)
t_lstm = time.time() - t0
print("LSTM (unpadded-all) training time (s):", t_lstm)
res_lstm = lstm_model.evaluate(pad_sequences(test_data, maxlen=maxlen_global, padding='post'),
                               y_test, verbose=0)
print("LSTM test loss/acc:", res_lstm)

# ========== УПОРЯДОЧИВАНИЕ ПО ДЛИНЕ + минимальное добавление нулей ==========
# 1) создаём список (sequence, label, length), сортируем по длине
train_pairs = list(zip(partial_x_train_seq, partial_y_train, [len(s) for s in partial_x_train_seq]))
train_pairs.sort(key=lambda x: x[2])  # сортируем по длине возрастанию

# 2) делаем minibatches так, чтобы в каждом батче минимальное расширение нулей
def make_length_sorted_batches(pairs, batch_size):
    batches = []
    for i in range(0, len(pairs), batch_size):
        batch = pairs[i:i+batch_size]
        seqs = [p[0] for p in batch]
        labels = np.array([p[1] for p in batch])
        maxlen = max(len(s) for s in seqs)
        seqs_padded = pad_sequences(seqs, maxlen=maxlen, padding='post')
        batches.append((seqs_padded, labels))
    return batches

batches = make_length_sorted_batches(train_pairs, BATCH_SIZE)
print("Number of batches (ordered):", len(batches))

# 3) Обучаем модель (создадим новую модель с mask_zero=True и будем обучать вручную по батчам)
lstm_model_ordered = keras.Sequential([
    layers.Embedding(NUM_WORDS, EMBED_DIM, input_length=None, mask_zero=True),
    layers.Bidirectional(layers.LSTM(LSTM_UNITS, dropout=0.1, recurrent_dropout=0.1)),
    layers.Dense(MAX_CLASSES, activation='softmax')
])
lstm_model_ordered.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
lstm_model_ordered.summary()

# Тренируем по батчам и измеряем время — для честного сравнения возьмём такое же число эпох:
t0 = time.time()
for epoch in range(EPOCHS_LSTM):
    for seqs_padded, labels in batches:
        lstm_model_ordered.train_on_batch(seqs_padded, labels)
    # можем опционально вычислить валидационную метрику на padded_x_val but padded to largest len in val:
    # пропустим для скорости
t_ordered = time.time() - t0
print("LSTM ordered-by-length training time (s):", t_ordered)

# Оценка ordered модели на тесте: подготовим тест, pad на максим длину батча для теста:
test_padded_for_eval = pad_sequences(test_data, padding='post')  # pad to global max
res_ordered = lstm_model_ordered.evaluate(test_padded_for_eval, y_test, verbose=0)
print("LSTM ordered test loss/acc:", res_ordered)

# ========== Сравнение времени и качества ==========
print("\n--- Summary ---")
print(f"Baseline time: {t_base:.1f}s, Baseline test acc: {res_base[1]:.4f}")
print(f"LSTM (all padded to same maxlen={maxlen_global}) time: {t_lstm:.1f}s, acc: {res_lstm[1]:.4f}")
print(f"LSTM (ordered, dynamic pad per-batch) time: {t_ordered:.1f}s, acc: {res_ordered[1]:.4f}")

# ========== Улучшения: ранняя остановка и регуляризация ==========
# Пример с EarlyStopping + L2 + Dropout
from tensorflow.keras.callbacks import EarlyStopping
early = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

model_improved = keras.Sequential([
    layers.Embedding(NUM_WORDS, EMBED_DIM, input_length=maxlen_global, mask_zero=True),
    layers.Bidirectional(layers.LSTM(128, return_sequences=False, dropout=0.2, recurrent_dropout=0.2)),
    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(1e-4)),
    layers.Dropout(0.3),
    layers.Dense(MAX_CLASSES, activation='softmax')
])
model_improved.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history_impr = model_improved.fit(padded_partial_x_train, partial_y_train,
                                  epochs=20, batch_size=64,
                                  validation_data=(padded_x_val, y_val),
                                  callbacks=[early], verbose=2)
print("Improved model test eval:", model_improved.evaluate(pad_sequences(test_data, maxlen=maxlen_global, padding='post'), y_test, verbose=0))

# ========== Plots (accuracy) ==========
plt.figure(figsize=(10,4))
plt.plot(history_lstm.history['accuracy'], 'bo-', label='LSTM train acc')
plt.plot(history_lstm.history['val_accuracy'], 'b.-', label='LSTM val acc')
if 'accuracy' in history_base.history:
    plt.plot(history_base.history['accuracy'], 'ro-', label='Baseline train acc')
    plt.plot(history_base.history['val_accuracy'], 'r.-', label='Baseline val acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
